# Import libraries

```python
import importlib
import numpy as np
from matplotlib import pyplot as plt

import specprop
from specprop.spectrum import (next_power_2, compute_complex_spectrum,
                      compute_spectral_amplitude, compute_power_spectrum, compute_psd)
```

```python
help(specprop)
```

# Generate signal

```python
n = 1000
fs = 100
dt = 1 / fs

f0 = 2
a = 3.0
phi = 30 * np.pi / 180

t = np.arange(n) * dt

x = a * np.sin(2.0 * np.pi * f0 * t + phi)

#--- add random noise
x += np.random.normal(0, 1, n)
```

# Plot signal

```python
plt.close("all")
fig = plt.figure(figsize=(10,3))
ax = fig.add_subplot()

ax.plot(t, x)

ax.set_xlabel('Time (s)')
ax.set_ylabel('Amplitude (unit)')
```

# Calculate spectral content

```python
f, A = compute_spectral_amplitude(x, fs)
f, P = compute_power_spectrum(x, fs)
f, PdB = compute_power_spectrum(x, fs, dB=True)
f, S = compute_psd(x, fs)
f, SdB = compute_psd(x, fs, taper=50, dB=True)
```
